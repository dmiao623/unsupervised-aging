```python
import os, json, functools
from pathlib import Path
import pandas as pd

feature_matrix_paths = [
    "2025-10-01_feature-matrix__combined_1126__2025-09-20_kpms-v5_150_1__2025-09-21_model-1.csv",
    "2025-10-01_feature-matrix__combined_1126__2025-09-20_kpms-v5_150_2__2025-09-21_model-1.csv",
    "2025-10-01_feature-matrix__combined_1126__2025-09-20_kpms-v5_150_3__2025-09-21_model-1.csv",
    "2025-10-06_feature-matrix__combined_1126__2025-09-20_kpms-v5_150_4__2025-09-21_model-1.csv",
    "2025-10-06_feature-matrix__combined_1126__2025-09-20_kpms-v5_150_5__2025-09-21_model-1.csv",
    "2025-10-06_feature-matrix__combined_1126__2025-09-20_kpms-v5_150_6__2025-09-21_model-1.csv",
    "2025-10-01_feature-matrix__combined_1126__2025-09-20_kpms-v5_150__old_1__2025-09-28_model-1.csv",
    "2025-10-01_feature-matrix__combined_1126__2025-09-20_kpms-v5_150__young_1__2025-09-28_model-1.csv",
]

xcat_json_paths = [
    "2025-10-01_xcats__combined_1126__2025-09-20_kpms-v5_150_1__2025-09-21_model-1.json",
    "2025-10-01_xcats__combined_1126__2025-09-20_kpms-v5_150_2__2025-09-21_model-1.json",
    "2025-10-01_xcats__combined_1126__2025-09-20_kpms-v5_150_3__2025-09-21_model-1.json",
    "2025-10-06_xcats__combined_1126__2025-09-20_kpms-v5_150_4__2025-09-21_model-1.json",
    "2025-10-06_xcats__combined_1126__2025-09-20_kpms-v5_150_5__2025-09-21_model-1.json",
    "2025-10-06_xcats__combined_1126__2025-09-20_kpms-v5_150_6__2025-09-21_model-1.json",
    "2025-10-01_xcats__combined_1126__2025-09-20_kpms-v5_150__old_1__2025-09-28_model-1.json",
    "2025-10-01_xcats__combined_1126__2025-09-20_kpms-v5_150__young_1__2025-09-28_model-1.json",
]

unsupervised_aging_dir = Path(os.environ["UNSUPERVISED_AGING"])
features_dir = unsupervised_aging_dir / "data/feature_matrices"

info = []
all_proj_cols = set()
supervised_cols = None
for csv_rel, json_rel in zip(feature_matrix_paths, xcat_json_paths):
    with open(features_dir / json_rel, "r") as f:
        x = json.load(f)
    for proj in [k for k in x if k != "supervised"]:
        proj_cols = x[proj]
        info.append((csv_rel, proj, proj_cols))
        all_proj_cols.update(proj_cols)
    if supervised_cols is None:
        supervised_cols = x["supervised"]

first_csv, first_proj, first_proj_cols = info[0]
df0 = pd.read_csv(features_dir / first_csv)
excluded = set(["name"]) | set(supervised_cols) | all_proj_cols
metadata_cols = [c for c in df0.columns if c not in excluded]
df0 = df0[["name"] + metadata_cols + supervised_cols + first_proj_cols]
df0 = df0.rename(columns={c: f"{first_proj}__{c}" for c in first_proj_cols})

dfs = [df0]
xcats_out = {first_proj: [f"{first_proj}__{c}" for c in first_proj_cols]}
for csv_rel, proj, proj_cols in info[1:]:
    df = pd.read_csv(features_dir / csv_rel, usecols=["name"] + proj_cols)
    df = df.rename(columns={c: f"{proj}__{c}" for c in proj_cols})
    dfs.append(df)
    xcats_out[proj] = [f"{proj}__{c}" for c in proj_cols]
xcats_out["supervised"] = supervised_cols

merged = functools.reduce(lambda l, r: pd.merge(l, r, on="name", how="inner"), dfs)

out_csv = "2025-10-06_feature-matrix__combined_1126__2025-09-20_kpms-v5_150_1_1-2-3-4-5-6-_old_1-_young-1.csv"
out_json = "2025-10-06_xcats__combined_1126__2025-09-20_kpms-v5_150_1_1-2-3-4-5-6-_old_1-_young-1.json"

merged.to_csv(features_dir / out_csv, index=False)
with open(features_dir / out_json, "w") as f:
    json.dump(xcats_out, f, indent=2)
```
